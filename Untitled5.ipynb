```python
import pandas as pd

# Create\ing a simple mock dataset (you can replace this with a CSV if you have one)
data = {
    'Patient_ID': [1, 2, 3, 4, 5],
    'Age': [45, 60, 29, 55, 70],
    'Gender': ['Male', 'Female', 'Female', 'Male', 'Female'],
    'Treatment_Type': ['Surgery', 'Medication', 'Therapy', 'Surgery', 'Therapy'],
    'Outcome': ['Recovery', 'Complications', 'Recovery', 'Recovery', 'No Improvement'],
    'Visit_Date': ['2022-05-01', '2022-06-12', '2022-07-25', '2022-08-30', '2022-09-15']
}

df=pd.DataFrame(data)
print(df['Outcome'].value_counts())
print(df['Treatment_Type'].value_counts())
```

```python
import matplotlib.pyplot as plt

# Plot age distribution
plt.hist(df['Age'], bins=5, color='skyblue', edgecolor='black')
plt.title('Age Distribution of Patients')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
import seaborn as sns

# Plot outcome by treatment type
plt.figure(figsize=(8,6))
sns.countplot(data=df, x='Treatment_Type', hue='Outcome', palette='Set2')
plt.title('Outcome by Treatment Type')
plt.xlabel('Treatment Type')
plt.ylabel('Number of Patients')
plt.show()
```

```python
gender_outcome = df.groupby(['Gender', 'Outcome']).size().unstack()
gender_outcome.plot(kind='pie', subplots=True, autopct='%1.1f%%', figsize=(8, 6), legend=False)
plt.title('Gender Distribution by Outcome')
plt.ylabel('')
plt.show()
```
